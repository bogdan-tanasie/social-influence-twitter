# Processing

Rename network related columns

```python
df.columns
```

```python
df.rename(columns={'A_network_feature_1': 'A_degree', 'A_network_feature_2': 'A_betweenness', 'A_network_feature_3': 'A_closeness',
                  'B_network_feature_1': 'B_degree', 'B_network_feature_2': 'B_betweenness', 'B_network_feature_3': 'B_closeness'}
          , inplace=True)
```

```python
df.columns
```

To facilitate the analysis (mainly condensing data while keeping information), we merge all columns. 

We will be using difference (A-B) and division (A/B) for this. Difference might be more interpretable, but as we do not have deep knowledge of the dataset, one format might be better than the other.
 
To allow of division, we will need to first replace 0 by 1. We replace by 1 in order to keep the data proportional. (A divison by 0.1 or lower might create very high numbers) We will only do this for graph that are not related to the graph and listed. (as these are fractions that can lead to infinity errors)

```python
# Replace zeros
df.loc[:, df.columns != 'Choice'].replace(0, 1)
```

```python
# Substraction
df['dAB_follower'] = df['A_follower_count'] - df['B_follower_count']
df['dAB_following'] = df['A_following_count'] - df['B_following_count']
df['dAB_listed'] = df['A_listed_count'] - df['B_listed_count']
df['dAB_mentions_received'] = df['A_mentions_received'] - df['B_mentions_received']
df['dAB_mentions_sent'] = df['A_mentions_sent'] - df['B_mentions_sent']
df['dAB_retweets_received'] = df['A_retweets_received'] - df['B_retweets_received']
df['dAB_retweets_sent'] = df['A_retweets_sent'] - df['B_retweets_sent']
df['dAB_posts'] = df['A_posts'] - df['B_posts']
df['dAB_degree'] = df['A_degree'] - df['B_degree']
df['dAB_betweenness'] = df['A_betweenness'] - df['B_betweenness']
df['dAB_closeness'] = df['A_closeness'] - df['B_closeness']

# Division 
df['rAB_follower'] = df['A_follower_count'] / df['B_follower_count']
df['rAB_following'] = df['A_following_count'] / df['B_following_count']
df['rAB_mentions_received'] = df['A_mentions_received'] / df['B_mentions_received']
df['rAB_mentions_sent'] = df['A_mentions_sent'] / df['B_mentions_sent']
df['rAB_retweets_received'] = df['A_retweets_received'] / df['B_retweets_received']
df['rAB_retweets_sent'] = df['A_retweets_sent'] / df['B_retweets_sent']
df['rAB_posts'] = df['A_posts'] / df['B_posts']
```

Drop the initial columns are redunant

```python
columns_to_drop = ['A_follower_count', 'B_follower_count', 'A_following_count', 'B_following_count', 'A_listed_count', 'B_listed_count',
          'A_mentions_received', 'B_mentions_received', 'A_mentions_sent', 'B_mentions_sent', 'A_retweets_received', 'B_retweets_received',
          'A_retweets_sent', 'B_retweets_sent', 'A_posts', 'B_posts']
df.drop(columns=columns_to_drop, inplace=True)
```

Separate tweets and non-tweets

```python
df.rename(columns={'dAB_posts': 'dAB_tweets', 'rAB_posts': 'rAB_tweets'}, inplace=True)
df['dAB_non_tweets_sent'] = df['dAB_mentions_sent'] + df['dAB_retweets_sent'] 
df['dAB_non_tweets_received'] = df['dAB_mentions_received'] + df['dAB_retweets_received']
df['rAB_non_tweets_sent'] = df['rAB_mentions_sent'] + df['rAB_retweets_sent'] 
df['rAB_non_tweets_received'] = df['rAB_mentions_received'] + df['rAB_retweets_received']
columns_to_drop = ['dAB_mentions_sent', 'dAB_retweets_sent', 'dAB_mentions_received', 'dAB_retweets_received',
                  'rAB_mentions_sent', 'rAB_retweets_sent', 'rAB_mentions_received', 'rAB_retweets_received']
df.drop(columns=columns_to_drop, inplace=True)
```
